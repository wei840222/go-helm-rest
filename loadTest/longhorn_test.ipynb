```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

上傳

```python
df = pd.read_csv('minio/longhorn_upload_test.csv')
df
```

```python
df.describe()
```

```python
points = 100

fig = plt.figure()
plt.xlabel('file')
plt.ylabel('time (sec)')
plt.plot(range(points), [df['4-MB'].iloc[x] for x in range(points)], '-', label='4-MB')
plt.plot(range(points), [df['8-MB'].iloc[x] for x in range(points)], '-', label='8-MB')
plt.plot(range(points), [df['16-MB'].iloc[x] for x in range(points)], '-', label='16-MB')
plt.legend(loc='upper left')
```

```python
points = 100

fig = plt.figure()
plt.xlabel('file')
plt.ylabel('time (sec)')
plt.plot(range(points), [df['64-MB'].iloc[x] for x in range(points)], '-', label='64-MB')
plt.plot(range(points), [df['128-MB'].iloc[x] for x in range(points)], '-', label='128-MB')
plt.plot(range(points), [df['256-MB'].iloc[x] for x in range(points)], '-', label='256-MB')
plt.legend(loc='upper left')
```
